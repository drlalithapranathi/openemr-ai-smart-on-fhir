```python
# Step 1: Load original CSV with transcripts
csv_path = './data/Medical Speech, Transcription, and Intent/overview-of-recordings.csv'
df_original = pd.read_csv(csv_path)

print(f"Loaded CSV: {len(df_original)} rows")
print(f"Columns: {list(df_original.columns)}\n")
```

```python
# Step 2: Get all audio files from VALIDATE directory
validate_dir = './data/Medical Speech, Transcription, and Intent/recordings/validate'

audio_files = []
for root, dirs, files in os.walk(validate_dir):
    for file in files:
        if file.endswith('.wav'):
            audio_files.append({
                'file_name': file,
                'filepath': os.path.join(root, file)
            })

print(f"Found {len(audio_files)} audio files in validate directory\n")
```

```python
# Step 4: Match with original transcripts using file_name
df_merged = df_audio.merge(
    df_original[['file_name', 'phrase', 'prompt', 'speaker_id']],
    on='file_name',
    how='left'
)

print(f"Matched {len(df_merged)} files with original transcripts")
print(f"Files without transcripts: {df_merged['phrase'].isna().sum()}\n")
print("First 5 matched files:")
print(df_merged[['file_name', 'phrase', 'prompt']].head())
```

```python
import torch
import torchaudio
import pandas as pd
import tempfile
import os

results = []

for i, row in df_merged.iterrows():
    audio_path = row['filepath']
    file_name = row['file_name']

    print(f"Transcribing file {i+1}/{len(df_merged)}: {file_name}")

    try:
        # Load the audio
        audio, sr = torchaudio.load(audio_path)

        # Convert stereo -> mono if needed
        if audio.shape[0] > 1:
            audio = torch.mean(audio, dim=0, keepdim=True)

        # Resample to 16kHz if needed
        if sr != 16000:
            audio = torchaudio.functional.resample(audio, sr, 16000)
            sr = 16000

        # Save temporary mono file
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file:
            temp_path = tmp_file.name
            torchaudio.save(temp_path, audio, sr)

        # Use temp file path for transcription
        answer_ids = model.generate(
            prompts=[
                [{"role": "user", "content": f"Transcribe the following: {model.audio_locator_tag}", "audio": [temp_path]}]
            ],
            max_new_tokens=128,
        )

        # Convert model output to text
        transcription = model.tokenizer.ids_to_text(answer_ids[0].cpu())

        # Remove temp file
        os.remove(temp_path)

    except Exception as e:
        transcription = f"Error: {e}"

    results.append({
        'file_name': file_name,
        'original_phrase': row['phrase'],
        'predicted_transcription': transcription,
        'speaker_id': row['speaker_id']
    })

# Save results
df_results = pd.DataFrame(results)
df_results.to_csv('canary_qwen_transcriptions.csv', index=False)
print("✅ Transcription complete!")
print(df_results.head())
```

```python
# --- Step 1: Install and import jiwer if not already ---
!pip install jiwer

import jiwer
import pandas as pd

# Load your dataframe with Qwen outputs
# Suppose your dataframe is called df_results and has columns:
# ['file_name', 'original_phrase', 'predicted_transcription', 'speaker_id']

print(f"Loaded {len(df_results)} rows")
print(f"Columns: {list(df_results.columns)}")

# --- Step 2: Create a jiwer transformation ---
transformation = jiwer.Compose([
    jiwer.RemovePunctuation(),
    jiwer.ToLowerCase(),
    jiwer.RemoveMultipleSpaces(),
    jiwer.Strip()
])

# --- Step 3: Normalize all references and hypotheses ---
refs = [transformation(str(r)) for r in df_results["original_phrase"]]       # Ground truth
hyps = [transformation(str(h)) for h in df_results["predicted_transcription"]]  # Qwen outputs

# --- Step 4: Calculate WER for each row ---
wer_scores = []
for i in range(len(refs)):
    try:
        wer_score = jiwer.wer(refs[i], hyps[i])
    except Exception as e:
        print(f"Error calculating WER for row {i}: {e}")
        wer_score = None
    wer_scores.append(wer_score)

# --- Step 5: Add WER scores to dataframe ---
df_results["wer"] = wer_scores

# --- Step 6: Calculate average WER (ignoring None) ---
valid_wer_scores = [w for w in wer_scores if w is not None]
average_wer = sum(valid_wer_scores) / len(valid_wer_scores)

# --- Step 7: Display results ---
print("\n=== RESULTS ===")
print(f"Model: nvidia/canary/qwen-2.5b")

print(f"\nFirst 5 WER scores:")
for idx, row in df_results.head(5).iterrows():
    print(f"  {row['file_name']}: {row['wer']:.4f}")

print(f"\nAverage WER: {average_wer:.4f}")

# --- Step 8: Save updated CSV with WER ---
output_path = "validate_qwen_transcripts_with_wer.csv"
df_results.to_csv(output_path, index=False)
print(f"\n✓ Saved CSV with WER scores to: {output_path}")

# --- Step 9: Optional: display summary table ---
print("\n=== SUMMARY TABLE (first 10 rows) ===")
print(df_results[["file_name", "predicted_transcription", "wer"]].head(10))
```

```python
import pandas as pd
import jiwer
from pathlib import Path

# --- Step 1: Load your CSV with Qwen transcriptions ---
csv_path = Path.cwd() / "validate_qwen_transcripts_with_wer.csv"  # Update with your actual file
df = pd.read_csv(csv_path)

print(f"Loaded {len(df)} rows")
print(f"Columns: {list(df.columns)}")

# --- Step 2: Define normalization transformation ---
transformation = jiwer.Compose([
    jiwer.RemovePunctuation(),
    jiwer.ToLowerCase(),
    jiwer.RemoveMultipleSpaces(),
    jiwer.Strip()
])

# --- Step 3: Generate word-level alignment visualizations ---
alignment_visualizations = []

for idx, row in df.iterrows():
    ref = transformation(str(row["original_phrase"]))            # Ground truth
    hyp = transformation(str(row["predicted_transcription"]))   # Qwen output

    try:
        # Process word-level alignment
        out = jiwer.process_words([ref], [hyp])
        alignment_str = jiwer.visualize_alignment(out)
    except Exception as e:
        print(f"Error at row {idx}: {e}")
        alignment_str = ""

    alignment_visualizations.append(alignment_str)

# --- Step 4: Add alignment visualization column to dataframe ---
df["alignment_visualization"] = alignment_visualizations

# --- Step 5: Save updated CSV ---
output_path = Path.cwd() / "wer_qwen2_5b_alignment.csv"
df.to_csv(output_path, index=False)

print("✅ Saved updated CSV with word alignment visualizations to:", output_path)

# --- Step 6: Display first few rows for sanity check ---
print("\n=== First 5 rows with alignment visualization ===")
print(df[["file_name", "original_phrase", "predicted_transcription", "alignment_visualization"]].head(5))
```

```python
import torch
import torchaudio
import pandas as pd
import tempfile
import os
import requests

# Load your CSV with downloadable audio URLs
csv_path = "capstone_audio.csv"  # Replace with your CSV path
df_merged = pd.read_csv(csv_path)

results = []

for i, row in df_merged.iterrows():
    file_name = row['diagnosis']
    audio_url = row['audio_download_url']  # Column with downloadable link
    print(f"Transcribing file {i+1}/{len(df_merged)}: {file_name}")

    try:
        # --- Step 1: Download the audio temporarily ---
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file:
            temp_download_path = tmp_file.name
            response = requests.get(audio_url)
            tmp_file.write(response.content)

        # --- Step 2: Load the audio ---
        audio, sr = torchaudio.load(temp_download_path)

        # Convert stereo -> mono if needed
        if audio.shape[0] > 1:
            audio = torch.mean(audio, dim=0, keepdim=True)

        # Resample to 16kHz if needed
        if sr != 16000:
            audio = torchaudio.functional.resample(audio, sr, 16000)
            sr = 16000

        # --- Step 3: Save as temp WAV for Qwen ---
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file2:
            temp_path = tmp_file2.name
            torchaudio.save(temp_path, audio, sr)

        # --- Step 4: Transcribe with Qwen ---
        answer_ids = model.generate(
            prompts=[
                [{"role": "user", "content": f"Transcribe the following: {model.audio_locator_tag}",
                  "audio": [temp_path]}]
            ],
            max_new_tokens=3000,
        )

        transcription = model.tokenizer.ids_to_text(answer_ids[0].cpu())

        # --- Step 5: Clean up temp files ---
        os.remove(temp_download_path)
        os.remove(temp_path)

    except Exception as e:
        transcription = f"Error: {e}"

    # --- Step 6: Append results ---
    results.append({
        'file_name': file_name,
        'original_phrase': row['ori_script'],
        'predicted_transcription': transcription,
        'speaker_id': row.get('speaker_id', None)  # Use None if no column
    })

# --- Step 7: Save all results ---
df_results = pd.DataFrame(results)
df_results.to_csv('canary_qwen_transcriptions_six.csv', index=False)
print("✅ Transcription complete!")
print(df_results.head())
```

```python
import torch
import torchaudio
import pandas as pd
import tempfile
import os
import time
import math

# Load your dataframe (update path)
df_merged = pd.read_csv("your_audio_metadata.csv")  # Should have columns: ['file_name', 'filepath', 'phrase', 'speaker_id']

results = []

# Set max_new_tokens for long transcription
MAX_NEW_TOKENS = 4500  # Adjust if needed
CHUNK_DURATION = 120  # seconds per chunk, 2 minutes

# Loop over audio files
for i, row in df_merged.iterrows():
    file_name = row['file_name']
    audio_path = row['filepath']
    print(f"\nTranscribing file {i+1}/{len(df_merged)}: {file_name}")

    start_time = time.time()
    transcription = ""

    try:
        # Load audio
        audio, sr = torchaudio.load(audio_path)

        # Convert stereo -> mono if needed
        if audio.shape[0] > 1:
            audio = torch.mean(audio, dim=0, keepdim=True)

        # Resample to 16kHz if needed
        if sr != 16000:
            audio = torchaudio.functional.resample(audio, sr, 16000)
            sr = 16000

        # Split audio into chunks
        num_chunks = math.ceil(audio.shape[1] / (CHUNK_DURATION * sr))
        for chunk_idx in range(num_chunks):
            start_sample = chunk_idx * CHUNK_DURATION * sr
            end_sample = min((chunk_idx + 1) * CHUNK_DURATION * sr, audio.shape[1])
            audio_chunk = audio[:, start_sample:end_sample]

            # Save temp chunk
            with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file:
                temp_path = tmp_file.name
                torchaudio.save(temp_path, audio_chunk, sr)

            # Transcribe chunk
            answer_ids = model.generate(
                prompts=[
                    [{"role": "user", "content": f"Transcribe the following: {model.audio_locator_tag}", "audio": [temp_path]}]
                ],
                max_new_tokens=MAX_NEW_TOKENS,
            )

            chunk_transcription = model.tokenizer.ids_to_text(answer_ids[0].cpu())
            transcription += " " + chunk_transcription.strip()

            # Remove temp file
            os.remove(temp_path)

    except Exception as e:
        transcription = f"Error: {e}"

    end_time = time.time()
    elapsed = end_time - start_time
    print(f"✅ Finished {file_name} in {elapsed:.1f} sec")

    results.append({
        'file_name': file_name,
        'original_phrase': row['phrase'],
        'predicted_transcription': transcription.strip(),
        'speaker_id': row['speaker_id'],
        'time_sec': elapsed
    })

# Save results to CSV
df_results = pd.DataFrame(results)
output_csv = "canary_qwen_transcriptions_with_time.csv"
df_results.to_csv(output_csv, index=False)

print(f"\n✅ All transcriptions complete! Saved to: {output_csv}")
print(df_results.head())
```
